```python
# static nodes and edges
nodeList = ["Cinderella",
	"Lady Tremaine",
	"Drizella Tremaine",
	"Anastasia Tremaine",
	"Jaq",
	"Gus",
	"Fairy Godmother",
	"Prince Charming"]

edgeList = [("Cinderella","Lady Tremaine"),
	("Lady Tremaine","Drizella Tremaine"),
	("Lady Tremaine","Anastasia Tremaine"),
	("Cinderella","Jaq"),
	("Cinderella","Gus"),
	("Cinderella","Fairy Godmother"),
	("Cinderella","Prince Charming")]

# networkx
import networkx as nx

# static data import and make network
G=nx.Graph()
G.add_nodes_from(nodeList)
G.add_edges_from(edgeList)
nx.draw(G, with_labels=True)


# matplotlib
import matplotlib.pyplot as plt
plt.show()


```
