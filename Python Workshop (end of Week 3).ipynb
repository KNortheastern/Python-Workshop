```python
# CSV nodes and edges
# pandas
import pandas as pd
nodeCSV = pd.read_csv('nodes.csv')
nodeCSVList = nodeCSV['Character'].values.tolist()
edgeCSV = pd.read_csv('edges.csv')
edgeCSVList = []
for row in edgeCSV.values:
    edgeCSVList.append(tuple(row))
    
# networkx
import networkx as nx

# csv data import and make network
G=nx.Graph()
G.add_nodes_from(nodeCSVList)
G.add_edges_from(edgeCSVList)
nx.draw(G, with_labels=True)


# matplotlib
import matplotlib.pyplot as plt
plt.show()
```
